Toxicity Dataset : https://archive.ics.uci.edu/dataset/728/toxicity-2

The dataset includes 171 molecules designed for functional domains of a core clock protein, CRY1, responsible for generating circadian rhythm. 56 of the molecules are toxic and the rest are non-toxic. 

The data consists a complete set of 1203 molecular descriptors and needs feature selection before classification since some of the features are redundant.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the CSV file
data = pd.read_csv("./data.csv")

# Display basic information about the dataset
print("Dataset shape:", data.shape)
```

```python
print("Column names:")
data.columns.tolist()
```

```python
# Separate features and target
# Assuming the last column or a column named 'Class' contains the target
if 'Class' in data.columns:
    X = data.drop('Class', axis=1)
    y = data['Class']
else:
    # Assume last column is the target
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
```

```python
# Basic data exploration
print("\n=== DATA EXPLORATION ===")
print(f"Shape of features (X): {X.shape}")
print(f"Shape of target (y): {y.shape}")

print(f"\nTarget distribution:")
print(y.value_counts())

print(f"\nClass balance:")
print(y.value_counts(normalize=True))

# Check target data type and unique values
print(f"\nTarget data type: {y.dtype}")
print(f"Unique target values: {y.unique()}")
```

```python
# Check for missing values
print(f"\nMissing values in features: {X.isnull().sum().sum()}")
print(f"Missing values in target: {y.isnull().sum()}")
```

```python
# Handle missing values if any
print("=== PREPROCESSING ===")

# Convert target to binary (1 for NonToxic, 0 for Toxic) - FLIPPED LABELS
y_binary = (y == 'NonToxic').astype(int)

# Verify the binary conversion
print("\nBinary target distribution:")
print(y_binary.value_counts())

print(f"\nClass balance: {y_binary.value_counts(normalize=True)}")

# Double-check the conversion is correct
print(f"\nMapping verification:")
print(f"Original 'NonToxic'  Binary {y_binary[y == 'NonToxic'].unique()}")
print(f"Original 'Toxic'  Binary {y_binary[y == 'Toxic'].unique()}")
```

```python
# Create X_filtered DataFrame from the variance-filtered data
X_filtered = pd.DataFrame(X, columns=X.columns)

# Split the data with stratification to ensure balanced folds
# Add some randomness to address potential ordering issues
np.random.seed(42)
shuffle_idx = np.random.permutation(len(X_filtered))
X_shuffled = X_filtered.iloc[shuffle_idx].reset_index(drop=True)
y_shuffled = y_binary.iloc[shuffle_idx].reset_index(drop=True)

X_train, X_test, y_train, y_test = train_test_split(
    X_shuffled, y_shuffled, test_size=0.2, random_state=42, 
    stratify=y_shuffled, shuffle=True
)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Training set size: {X_train_scaled.shape}")
print(f"Test set size: {X_test_scaled.shape}")

# Check class distribution in train and test sets
print(f"\nTrain set class distribution:")
print(pd.Series(y_train).value_counts(normalize=True))
print(f"\nTest set class distribution:")
print(pd.Series(y_test).value_counts(normalize=True))
```

```python
# Define evaluation function
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    """Evaluate a classification model and return metrics"""
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Probabilities for AUC
    if hasattr(model, 'predict_proba'):
        y_train_proba = model.predict_proba(X_train)[:, 1]
        y_test_proba = model.predict_proba(X_test)[:, 1]
    else:
        y_train_proba = model.decision_function(X_train)
        y_test_proba = model.decision_function(X_test)
    
    # Calculate metrics
    metrics = {
        'model': model_name,
        'train_accuracy': accuracy_score(y_train, y_train_pred),
        'test_accuracy': accuracy_score(y_test, y_test_pred),
        'train_auc': roc_auc_score(y_train, y_train_proba),
        'test_auc': roc_auc_score(y_test, y_test_proba),
        'precision': precision_score(y_test, y_test_pred),
        'recall': recall_score(y_test, y_test_pred),
        'f1': f1_score(y_test, y_test_pred)
    }
    
    return metrics, y_test_pred, y_test_proba
```

```python
# Initialize results storage
results = []
all_predictions = {}
all_probabilities = {}
```

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression

# Baseline Logistic Regression with RFECV (no regularization)
rfecv = RFECV(LogisticRegression(penalty=None, max_iter=2000, solver='lbfgs'), step=1, cv=5, scoring='accuracy')
rfecv.fit(X_train_scaled, y_train)
X_train_rfecv = rfecv.transform(X_train_scaled)
X_test_rfecv = rfecv.transform(X_test_scaled)
ordinary_lr = rfecv.estimator_
ordinary_lr.fit(X_train_rfecv, y_train)
ordinary_metrics, ordinary_pred, ordinary_proba = evaluate_model(
    ordinary_lr, X_train_rfecv, X_test_rfecv, y_train, y_test, 'Ordinary LR (RFECV)'
)
results.append(ordinary_metrics)
all_predictions['Ordinary LR (RFECV)'] = ordinary_pred
all_probabilities['Ordinary LR (RFECV)'] = ordinary_proba
print(f"Optimal number of features: {rfecv.n_features_}")
```

```python
# Lasso Logistic Regression (L1)
lasso = LogisticRegression(penalty='l1', solver='lbfgs', max_iter=2000)

lasso_rfecv = RFECV(
    estimator=lasso,
    step=1,
    cv=5,
    scoring='accuracy'
)
lasso_rfecv.fit(X_train_scaled, y_train)
X_train_lasso_rfecv = lasso_rfecv.transform(X_train_scaled)
X_test_lasso_rfecv = lasso_rfecv.transform(X_test_scaled)
lasso.fit(X_train_lasso_rfecv, y_train)
```

```python
# Ridge Logistic Regression (L2) with RFECV
ridge = LogisticRegression(penalty='l2', max_iter=2000, solver='lbfgs')
ridge_rfecv = RFECV(estimator=ridge, step=1, cv=5, scoring='accuracy')
ridge_rfecv.fit(X_train_scaled, y_train)
X_train_rfecv = ridge_rfecv.transform(X_train_scaled)
X_test_rfecv = ridge_rfecv.transform(X_test_scaled)
ridge.fit(X_train_rfecv, y_train)
ridge_metrics, ridge_pred, ridge_proba = evaluate_model(
    ridge, X_train_rfecv, X_test_rfecv, y_train, y_test, 'Ridge LR (RFECV)'
)
results.append(ridge_metrics)
all_predictions['Ridge LR (RFECV)'] = ridge_pred
all_probabilities['Ridge LR (RFECV)'] = ridge_proba
print(f"Optimal features: {ridge_rfecv.n_features_}")
print("Feature ranking:", ridge_rfecv.ranking_)
```

```python
# Elastic Net Logistic Regression
elasticnet = LogisticRegression(penalty='elasticnet', solver='saga', l1_ratio=0.5, max_iter=2000)

elasticnet_rfecv = RFECV(
    estimator=elasticnet,
    step=1,
    cv=5,
    scoring='accuracy'
)
elasticnet_rfecv.fit(X_train_scaled, y_train)
X_train_elastic_rfecv = elasticnet_rfecv.transform(X_train_scaled)
X_test_elastic_rfecv = elasticnet_rfecv.transform(X_test_scaled)
elasticnet.fit(X_train_elastic_rfecv, y_train)
```
